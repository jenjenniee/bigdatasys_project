```python
import pandas as pd
import numpy as np

df = pd.read_csv('D:/4th/bigdatasys/bsd-project/bigdatasys-proj/Random_Delivery_Data_with_Neighborhoods.csv')

df.head()
```

```python
df['DO'] = np.nan
df['SI'] = np.nan
df['GU'] = np.nan
df['DONG'] = np.nan
```

```python
df
```

```python
def categorize_area(row):
    parts = row['AREA'].split()
    for part in parts:
        if part.endswith('도'):
            row['DO'] = part
        elif part.endswith('시'):
            row['SI'] = part
        elif part.endswith('구'):
            row['GU'] = part
        elif part.endswith('동'):
            row['DONG'] = part
    return row
```

```python
df = df.apply(categorize_area, axis=1)

df
```

```python
df.drop('AREA', axis=1, inplace=True)
```

```python
df.to_csv('delivery.csv', index=False)
```
